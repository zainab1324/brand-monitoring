# Brand Monitoring Analysis
This notebook loads the available brand monitoring dataset, performs sentiment and topic analysis, and produces actionable insights aligned with the project objectives.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
from wordcloud import WordCloud
import nltk
from nltk.corpus import stopwords

print("pandas", pd.__version__)
print("numpy", np.__version__)
print("plotly", px.__version__)
print("wordcloud ready")
```

## Data Loading and Exploration
Load the tweet dataset and inspect data quality, sentiment labels, and time fields.

```python
tweets = pd.read_csv("../data/Tweets.csv", parse_dates=["tweet_created"], infer_datetime_format=True)
tweets["text"] = tweets["text"].astype(str)
tweets["tweet_date"] = pd.to_datetime(tweets["tweet_created"], errors="coerce").dt.date
tweets.head()
```

```python
tweets.info()
print("Missing values by column:
", tweets.isna().sum())
```

## Sentiment Analysis
Review the distribution of sentiment across brand mentions and identify the dominant mood drivers.

```python
sentiment_counts = tweets["airline_sentiment"].value_counts()
fig = px.bar(x=sentiment_counts.index, y=sentiment_counts.values, labels={"x":"Sentiment","y":"Count"}, title="Sentiment distribution")
fig.show()
```

## Topic Modeling and Emerging Themes
Extract the top terms and build a word cloud to surface common discussion themes.

```python
nltk.download("stopwords")
stop_words = set(stopwords.words("english"))
custom_stop = {"virginamerica","virgin","america","flight","http","https","t.co","rt"}
tokens = [word.lower() for word in " ".join(tweets["text"]).split() if len(word) > 3]
filtered = [word for word in tokens if word.isalpha() and word not in stop_words and word not in custom_stop]
wordcloud = WordCloud(width=1000, height=500, background_color="white").generate(" ".join(filtered))
plt.figure(figsize=(14, 7))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

## Trend Analysis
Analyze how sentiment changes over time to identify potential early-warning signals.

```python
trend = tweets.groupby(["tweet_date", "airline_sentiment"]).size().reset_index(name="count")
fig = px.area(trend, x="tweet_date", y="count", color="airline_sentiment", labels={"tweet_date":"Date","count":"Mentions"}, title="Daily sentiment trend")
fig.show()
```

## Actionable Insights
Summarize impact, issue escalation, and next-step recommendations for brand management.
